# About Data Set
Zomato is one of the most-used apps to find food and restaurants in and around one locality to order and check reviews. Zomato has not only solved the problem of ordering food sitting at your home but also has provided a single platform to know more details about each restaurant. This tabular dataset consists of all restaurants in Bangalore. Most of the people here are dependent mainly on the restaurant food as they don’t have time to cook for themselves.


### Column Description
•	Name : Name of resturants<br>
•	Cuisines : related resturant food<br>
•	Area : Area of resturant<br>
•	Addresss : Complete addresss in related area<br>
•	Phone Number : Phone number of resturant<br>
•	isHomeDelivery : is related food is home deliver order or not<br>
•	is take away : ...<br>
•	isIndoorSeating : ...<br>
•	isVegOnly : 0 mean not have vegitable and 1 have<br>
•	Dinner Rating : related food rating as dinner time<br>
•	Lunch Rating : related food rating as lunch time<br>
•	Delivery Rating : Rating of delivery<br>
•	Delivery Review : Number of orders<br>
•	KnownFor : Why it is famous<br>
•	PopularDishes : Popular Dishes<br>
•	Cost : Price of food<br>

From all the Data available, we can bring out some neat insights or conclusions such as :<br>
•	What is the most liked Dish type?<br>
•	How many types of Restaurant types are there?<br>
•	What is the most liked Restaurant type?<br>
•	What is the Average cost of food?<br>
•	Which franchise is the most popular and which has the highest no of rating?<br>
•	Is there food deliver or not?<br>
•	Does location affect the rating?<br>
And so….

### Mapping DataFrame

```python
def map_rating_to_category(Rating):
    if Rating>'3.7':
        return "High Rating"
    elif Rating<'2':
        return "Low Rating"
    else:
        return "Medium Rating"
```
